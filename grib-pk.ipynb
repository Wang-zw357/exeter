```python
import os
import zipfile
 
def zipDir(dirpath, outFullName):

    zip = zipfile.ZipFile(outFullName, "w", zipfile.ZIP_DEFLATED)
    for path, dirnames, filenames in os.walk(dirpath):

        fpath = path.replace(dirpath, '')
 
        for filename in filenames:
            zip.write(os.path.join(path, filename), os.path.join(fpath, filename))
    zip.close()
 
 

input_path = "rawdata"
output_path = "rawdata.zip"

zipDir(input_path, output_path)
```

```python
LAT1 = None
LAT2 = None
LON1 = None
LON2 = None

def process_message(grb, outfile):
    tmps = "%s" % grb
    wtitle = tmps.split(':')[1]
    wtime = tmps.split(':')[len(tmps.split(':'))-1].split(' ')[1]
    data, lat, lons = grb.data(LAT1, LAT2, LON1, LON2)
    for i in range(len(data)):
        tdata = data[i]
        tlat = lat[i]
        tlon = lons[1]
        for j in range(len(tdata)):
            wdata = tdata[j]
            wlat = tlat[j]
            wlon = tlon[j]
            outfile.write("%s, %s, %s, %s, %s\n" % (wtitle, wtime, wlat, wlon, wdata)) 
```
